```python
import pandas as pd
def reassign(row):
    try:
        row=str(row).strip()
        return float(row)
    except(ValueError):
        

        if row.endswith('Sq. Meter'):
            x=float(row.split('Sq. Meter')[0].strip())
            row= x*10.76
            return row
        elif row.endswith('Sq. Yards'): 
            x=float(row.split('Sq. Yards')[0].strip())
            row=x*9
            return row
        elif row.endswith('Acres'):
            x=float(row.split('Acres')[0].strip())
            row=x*43560
            return row
            
        elif "-" in row:
            x=[float(p.strip()) for p in row.split("-")]
            return sum(x)/2
        
        

data=pd.read_csv("bengaluru_house_prices.csv")
data=data.dropna()
data['total_sqft']=data['total_sqft'].apply(reassign)
```
